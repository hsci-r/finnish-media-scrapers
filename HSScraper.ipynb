```python
# Hae

from bs4 import BeautifulSoup
#import urllib2
import re

driver.get("https://www.hs.fi/haku/?query=SDP&category=kaikki&period=custom&order=new&startDate=2020-07-01&endDate=2020-12-31")


links = set()

previous = 0

while True:
    elements = driver.find_elements_by_css_selector('.w-full .max-w-main')
    for element in elements:
        html = element.get_attribute('innerHTML')
        soup = BeautifulSoup(html)
        for link in soup.findAll('a'):
            #print(link.get('href'))
            links.add(link.get('href'))
        #link = driver.find_elements_by_css_selector("a .block")[0]
        #print(link.get_attribute('href'))
    driver.find_elements_by_xpath("//*[contains(text(), 'Nyt lis')]")[0].click()
    time.sleep(2)
    if len(links) == previous:
        break
    else:
        previous = len(links)
    print(len(links))
```

```python
    
f = open("HS_links.txt", "w+")

f.write("\n".join(links))
f.close()


#driver.close()
```
